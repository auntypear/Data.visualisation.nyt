```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv("nytimes_best_restaurants_2024.csv", encoding="ISO-8859-1")

# Drop rows where all elements are NaN (fully empty rows)
df_cleaned = df.dropna(how='all')

# Save the cleaned DataFrame to a new CSV
df_cleaned.to_csv('cleaned_restaurants.csv', index=False)

# Save the state counts as a DataFrame and then to a CSV
state = df_cleaned['state'].value_counts()
state.to_csv('state_counts.csv', header=['Count'], index_label='State')

# Convert the Series into a DataFrame for plotting
state_df = state.reset_index()
state_df.columns = ['State', 'Count']

# Create a bar plot
plt.figure(figsize=(12, 6))  # Adjust the figure size as needed
sns.barplot(x='State', y='Count', data=state_df, palette='twilight', hue='State', dodge=False, legend=False)

# Set the title and labels
plt.title('States')
plt.xlabel('State')
plt.ylabel('Number of Restaurants')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Adjust layout to avoid clipping
plt.tight_layout()

# Show the plot
plt.show()
```
